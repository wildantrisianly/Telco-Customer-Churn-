# Latar Belakang

Industri telekomunikasi merupakan salah satu sektor dengan tingkat persaingan yang sangat tinggi. Perusahaan penyedia layanan seperti internet, WiFi, dan telepon harus terus berinovasi, tidak hanya dalam meningkatkan kualitas jaringan, tetapi juga dalam mempertahankan pelanggan yang sudah ada. Salah satu indikator utama yang sering digunakan untuk mengukur keberhasilan strategi retensi pelanggan adalah **churn rate**, yaitu persentase pelanggan yang berhenti menggunakan layanan dalam periode tertentu (Investopedia, 2023).

Churn rate menjadi permasalahan krusial dalam industri telekomunikasi karena pelanggan cenderung bersifat **price-sensitive** dan memiliki banyak alternatif penyedia layanan. Kemudahan berpindah operator, perbedaan harga paket, serta kualitas layanan yang dirasakan pelanggan menjadi faktor utama yang mendorong terjadinya churn. Berdasarkan laporan industri, tingkat churn pada sektor telekomunikasi secara global berada pada kisaran **15% hingga 25% per tahun**, tergantung pada jenis layanan dan segmentasi pelanggan (Statista, 2022).

Tingginya churn rate dapat memberikan dampak negatif yang signifikan terhadap kinerja bisnis perusahaan. Kehilangan pelanggan tidak hanya berarti berkurangnya pendapatan, tetapi juga meningkatkan biaya operasional akibat kebutuhan untuk mengakuisisi pelanggan baru. Beberapa studi menyebutkan bahwa **biaya untuk mendapatkan pelanggan baru dapat mencapai lima hingga sepuluh kali lebih besar dibandingkan mempertahankan pelanggan yang sudah ada** (Harvard Business Review, 2014).

Dalam konteks ini, perusahaan seperti **Telco Technology** yang menyediakan layanan internet, WiFi, dan telepon bagi pelanggan individu maupun bisnis perlu memiliki strategi yang efektif untuk mengurangi tingkat churn. Pendekatan konvensional yang bersifat reaktif, yaitu bertindak setelah pelanggan berhenti berlangganan, sering kali kurang optimal dalam menekan churn rate.

Seiring dengan meningkatnya ketersediaan data pelanggan dan perkembangan teknologi analitik, **machine learning** menjadi pendekatan yang potensial untuk mengatasi permasalahan churn. Dengan memanfaatkan data historis pelanggan, model machine learning dapat digunakan untuk **memprediksi pelanggan yang berisiko churn** berdasarkan pola perilaku, penggunaan layanan, dan karakteristik pelanggan lainnya. Pendekatan prediktif ini memungkinkan perusahaan untuk mengambil tindakan proaktif dalam strategi retensi pelanggan (IBM, 2023).

Oleh karena itu, penelitian ini berfokus pada penerapan metode **machine learning untuk prediksi churn pelanggan** pada perusahaan telekomunikasi, dengan tujuan membantu pengambilan keputusan bisnis berbasis data serta meningkatkan loyalitas pelanggan dalam jangka panjang.

## **Metric Evaluation**

Dalam evaluasi model klasifikasi churn, digunakan **confusion matrix** untuk memahami performa model dalam memprediksi pelanggan yang churn dan tidak churn. Confusion matrix membandingkan hasil prediksi model dengan kondisi aktual pelanggan, yang terdiri dari empat komponen utama:

- **True Negative (TN)**  
  Model memprediksi pelanggan **tidak churn (0)** dan hasil tersebut sesuai dengan kondisi aktual.

- **False Positive (FP)**  
  Model memprediksi pelanggan **churn (1)**, padahal pada kenyataannya pelanggan **tidak churn**.

- **False Negative (FN)**  
  Model memprediksi pelanggan **tidak churn (0)**, padahal pelanggan tersebut **benar-benar churn**.

- **True Positive (TP)**  
  Model memprediksi pelanggan **churn (1)** dan hasil tersebut sesuai dengan kondisi aktual.

### Type of Errors

Dalam konteks churn prediction, terdapat dua jenis kesalahan utama yang perlu diperhatikan:

- **Type I Error (False Positive)**  
  Terjadi ketika model memprediksi pelanggan akan churn, namun sebenarnya pelanggan tersebut tidak churn.  
  Kesalahan ini dapat menyebabkan perusahaan mengambil tindakan retensi yang tidak diperlukan, sehingga berpotensi menimbulkan **pemborosan sumber daya**.

- **Type II Error (False Negative)**  
  Terjadi ketika model memprediksi pelanggan tidak churn, padahal sebenarnya pelanggan tersebut akan churn.  
  Kesalahan ini lebih berisiko karena perusahaan **kehilangan peluang untuk mempertahankan pelanggan**, yang dapat berdampak langsung pada **kehilangan pendapatan**.


**Pertimbangan Cost-Benefit dalam Prediksi Churn Telco**

Dalam industri telekomunikasi, kesalahan prediksi pada model churn tidak memiliki dampak bisnis yang sama. Secara khusus, kesalahan dalam bentuk False Negative (FN) memiliki konsekuensi bisnis yang jauh lebih besar dibandingkan False Positive (FP).

False Negative terjadi ketika model memprediksi pelanggan tidak churn, namun pada kenyataannya pelanggan tersebut benar-benar churn. Kondisi ini menyebabkan perusahaan kehilangan pelanggan tanpa adanya intervensi, yang berdampak pada hilangnya pendapatan berulang serta meningkatnya biaya akuisisi pelanggan baru. Sebaliknya, False Positive terjadi ketika model memprediksi pelanggan akan churn padahal sebenarnya tidak, yang umumnya hanya menimbulkan biaya tambahan berupa insentif atau program retensi yang tidak diperlukan.

Nilai biaya yang digunakan tidak dimaksudkan sebagai angka absolut, melainkan sebagai pendekatan yang realistis dan defensible secara bisnis. Dalam konteks industri telekomunikasi, biaya kehilangan pelanggan umumnya berada pada kisaran **5 hingga 10 kali lebih besar** dibandingkan biaya retensi pelanggan.

Pendekatan ini selaras dengan praktik cost-sensitive learning dan menjadi dasar dalam pemilihan metrik evaluasi model yang lebih menekankan pada kemampuan model dalam menangkap pelanggan yang benar-benar churn.


**Implikasi Bisnis terhadap Evaluasi Model**

Berdasarkan analisis cost-benefit di atas, dapat disimpulkan bahwa False Negative merupakan jenis kesalahan yang paling merugikan bagi perusahaan. Kehilangan satu pelanggan memiliki dampak finansial yang jauh lebih besar dibandingkan biaya intervensi yang tidak diperlukan pada pelanggan yang sebenarnya loyal.

Oleh karena itu, dalam konteks prediksi churn, model harus lebih diprioritaskan untuk meminimalkan False Negative, meskipun hal tersebut dapat meningkatkan jumlah False Positive. Dengan kata lain, kemampuan model dalam menangkap sebanyak mungkin pelanggan yang benar-benar churn menjadi lebih penting dibandingkan sekadar memaksimalkan akurasi secara keseluruhan.

**Alasan Pemilihan F2-Score**

Untuk mendukung tujuan bisnis tersebut, metrik evaluasi yang digunakan dalam proyek ini adalah F2-Score. Berbeda dengan F1-Score yang memberikan bobot seimbang antara precision dan recall, F2-Score memberikan bobot lebih besar pada recall.

Dengan menekankan recall, F2-Score membantu memastikan bahwa model mampu mengidentifikasi sebanyak mungkin pelanggan yang berpotensi churn, sehingga risiko False Negative dapat diminimalkan. Pendekatan ini lebih selaras dengan kebutuhan bisnis perusahaan telekomunikasi, di mana kehilangan pelanggan memiliki dampak finansial yang signifikan.

### Metric Selection

Dalam permasalahan churn, **False Negative** dianggap lebih kritis dibandingkan False Positive. Oleh karena itu, tujuan utama dari model adalah **meminimalkan False Negative**, sehingga pelanggan yang benar-benar berisiko churn dapat terdeteksi lebih awal.

Untuk mencapai tujuan tersebut, metrik evaluasi utama yang digunakan adalah **F2-Score**, karena metrik ini memberikan bobot yang lebih besar pada **Recall** dibandingkan Precision. Dengan demikian, model lebih difokuskan pada kemampuan mendeteksi pelanggan churn sebanyak mungkin.

### F2-Score Formula


$$F2 = \frac{(1 + 2^2) \times (Precision \times Recall)}{(2^2 \times Precision) + Recall}$$


Dimana:

- **Precision**  
\[
Precision = $\frac{TP}{TP + FP}$
\]

- **Recall**  
\[
Recall = $\frac{TP}{TP + FN}$
\]

Keterangan:
- **TP (True Positive)**: Jumlah pelanggan churn yang diprediksi dengan benar  
- **FP (False Positive)**: Jumlah pelanggan tidak churn yang salah diprediksi churn  
- **FN (False Negative)**: Jumlah pelanggan churn yang salah diprediksi tidak churn  

Dengan menggunakan **F2-Score** sebagai metrik utama, model diharapkan mampu mengidentifikasi pelanggan yang berisiko churn secara lebih optimal, sehingga perusahaan dapat mengalokasikan strategi retensi secara lebih efektif dan mengurangi risiko kehilangan pelanggan.

## **Preprocess**

```python
# 1. Pisahkan tipe kolom
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = X_train.select_dtypes(include=['object', 'bool']).columns.tolist()

print("Numeric:", numeric_cols)
print("Categorical:", categorical_cols)


# 2. Proses kolom numerik
numeric_process = Pipeline([
    ('fill_na', SimpleImputer(strategy='median')),   # isi missing value
    ('scale', RobustScaler())                        # scaling angka
])


# 3. Proses kolom kategori
categorical_process = Pipeline([
    ('fill_na', SimpleImputer(strategy='most_frequent')),  # isi missing value
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'))  # ubah ke angka
])


# 4. Proses kolom onehot 
onehot_process = Pipeline([
    ('fill_na', SimpleImputer(strategy='most_frequent')),
    ('onehot', OrdinalEncoder(categories=[[1,2,3,4,5]]))
])


# 5. Gabungkan semua
preprocessor = ColumnTransformer([
    ('num', numeric_process, numeric_cols),
    ('onehot', onehot_process, ['satisfactionscore']),
    ('cat', categorical_process, categorical_cols)
])


# 6. Transform data
X_train_final = preprocessor.fit_transform(X_train)
X_test_final  = preprocessor.transform(X_test)

print("Shape setelah preprocessing:", X_train_final.shape)
```

## Preprocessing Insight

Untuk memastikan pipeline machine learning lebih rapi dan mengurangi risiko **data leakage**, preprocessing dilakukan menggunakan **ColumnTransformer**. Proses `fit_transform` hanya dilakukan pada data train, sedangkan data test hanya menggunakan `transform`.

### Fitur Numerik

1. Missing value pada fitur numerik diisi menggunakan **median**.
2. Median digunakan karena lebih robust terhadap outlier dibandingkan mean.
3. Setelah itu, fitur numerik diskalakan menggunakan **RobustScaler**.
4. RobustScaler digunakan untuk mengurangi pengaruh outlier pada data numerik.

### Fitur Kategorikal

1. Missing value pada fitur kategorikal diisi menggunakan nilai yang paling sering muncul atau **modus**.
2. Fitur kategorikal kemudian dikodekan menggunakan **One-Hot Encoding** agar dapat dibaca oleh model machine learning.
3. Parameter `handle_unknown='ignore'` digunakan agar pipeline tetap aman jika terdapat kategori baru pada data test.
4. Parameter `drop='first'` digunakan untuk mengurangi redundansi antar kategori hasil encoding.

### Fitur Ordinal

1. Fitur **satisfactionscore** diperlakukan sebagai fitur ordinal karena memiliki urutan nilai dari **1 sampai 5**.
2. Fitur ini dikodekan menggunakan **OrdinalEncoder** agar urutan tingkat kepuasan pelanggan tetap dipertahankan.
3. Pendekatan ini lebih sesuai dibanding One-Hot Encoding karena nilai satisfaction score memiliki makna tingkatan.

# Cost-Benefit Analysis

Untuk mengevaluasi dampak bisnis dari model, digunakan pendekatan cost-benefit berdasarkan jenis kesalahan prediksi.

Asumsi biaya:

- **False Positive (FP): $20**  
  Biaya intervensi retention yang tidak tepat sasaran, misalnya promo atau diskon kepada customer yang sebenarnya tidak churn.

- **False Negative (FN): $200**  
  Biaya kehilangan customer churn yang tidak terdeteksi, seperti kehilangan revenue, customer lifetime value, dan biaya akuisisi customer baru.

---

## 1. Tanpa Model

Sebelum menggunakan model machine learning, perusahaan dapat menggunakan pendekatan manual atau strategi umum. Untuk menggambarkan kondisi ini, digunakan dua skenario sederhana:

### Skenario Pesimis: Semua Customer Diprediksi Churn

Pada skenario ini, perusahaan menganggap semua customer berisiko churn, sehingga semua customer diberikan intervensi retention.

Berdasarkan data test:

- Actual churn customer = 374
- Actual not churn customer = 1,035

Jika semua customer diprediksi churn:

- 374 customer churn berhasil ditargetkan dengan benar
- 1,035 customer yang sebenarnya tidak churn tetap diberikan intervensi

Perhitungan cost:

- FP = 1,035
- Cost = 1,035 × $20 = **$20,700**

Artinya, strategi ini terlalu agresif karena perusahaan mengeluarkan biaya retention kepada banyak customer yang sebenarnya tidak churn.

---

### Skenario Optimis: Semua Customer Diprediksi Tidak Churn

Pada skenario ini, perusahaan tidak melakukan intervensi retention kepada customer mana pun karena menganggap semua customer tidak akan churn.

Jika semua customer diprediksi tidak churn:

- 1,035 customer not churn tidak membutuhkan intervensi
- 374 customer churn tidak terdeteksi dan tidak mendapatkan treatment retention

Perhitungan cost:

- FN = 374
- Cost = 374 × $200 = **$74,800**

Artinya, strategi ini sangat berisiko karena perusahaan melewatkan seluruh customer yang benar-benar churn.

---

## 2. Before Tuning: Base Model

Berdasarkan confusion matrix Base Model:

- True Negative (TN) = 973
- False Positive (FP) = 62
- False Negative (FN) = 21
- True Positive (TP) = 353

Perhitungan cost:

- FP cost = 62 × $20 = **$1,240**
- FN cost = 21 × $200 = **$4,200**

Total cost Base Model:

**$1,240 + $4,200 = $5,440**

---

## 3. After Tuning: Best Model

Berdasarkan confusion matrix Best Model yang sudah menggunakan best parameter dan threshold:

- True Negative (TN) = 941
- False Positive (FP) = 94
- False Negative (FN) = 8
- True Positive (TP) = 366

Perhitungan cost:

- FP cost = 94 × $20 = **$1,880**
- FN cost = 8 × $200 = **$1,600**

Total cost Best Model:

**$1,880 + $1,600 = $3,480**

---

## 4. Cost Reduction

Setelah dilakukan tuning parameter dan threshold optimization, total cost menurun dari Base Model ke Best Model.

Cost saving:

**$5,440 - $3,480 = $1,960**

Artinya, Best Model mampu menurunkan estimasi biaya sebesar **$1,960** dibandingkan Base Model.

Jika dibandingkan dengan skenario tanpa model:

| Tanpa Model - Semua Diprediksi Tidak Churn | $74,800 |

## Insight

Berdasarkan hasil cost-benefit analysis, penggunaan machine learning model memberikan dampak bisnis yang lebih optimal dibandingkan strategi tanpa model.

Pada skenario tanpa model, strategi yang terlalu agresif dengan menargetkan semua customer sebagai churn menghasilkan biaya sebesar **$20,700**, karena banyak customer yang sebenarnya tidak churn tetap diberikan intervensi retention. Sebaliknya, strategi yang terlalu pasif dengan menganggap semua customer tidak churn menghasilkan biaya yang jauh lebih besar, yaitu **$74,800**, karena seluruh customer churn tidak terdeteksi.

Best Model memang menghasilkan False Positive yang lebih tinggi dibanding Base Model, yaitu dari **62 menjadi 94**. Namun, model juga berhasil menurunkan False Negative secara signifikan dari **21 menjadi 8**. Dalam konteks churn prediction, penurunan False Negative lebih penting karena customer churn yang tidak terdeteksi memiliki biaya kerugian lebih besar.

Secara keseluruhan, Best Model dipilih sebagai final model karena mampu meningkatkan kemampuan deteksi customer churn, menurunkan total cost sebesar **$1,960** dibanding Base Model, dan memberikan potensi cost saving hingga **$71,320** dibandingkan skenario tanpa model yang tidak melakukan intervensi.

## Business Recommendations

Berdasarkan hasil EDA, model evaluation, threshold optimization, dan SHAP interpretation, terdapat beberapa rekomendasi bisnis yang dapat dilakukan perusahaan untuk menurunkan risiko customer churn:

### 1. Prioritaskan customer dengan satisfaction score rendah

Hasil SHAP menunjukkan bahwa **satisfactionscore** menjadi fitur paling berpengaruh terhadap prediksi churn. Customer dengan tingkat kepuasan rendah memiliki kemungkinan churn yang lebih tinggi.

**Recommendation:**
Perusahaan perlu membuat sistem monitoring terhadap customer dengan satisfaction score rendah. Customer pada segmen ini dapat diprioritaskan untuk program retention seperti follow-up keluhan, service recovery, pemberian promo khusus, atau peningkatan kualitas customer support.


### 2. Tawarkan bundling layanan tambahan seperti Online Security

Fitur **onlinesecurity** juga menjadi salah satu faktor penting dalam prediksi churn. Customer yang tidak menggunakan layanan Online Security cenderung memiliki risiko churn lebih tinggi.

**Recommendation:**
Perusahaan dapat menawarkan paket bundling internet dengan Online Security, Tech Support, atau layanan tambahan lain. Tujuannya adalah meningkatkan perceived value customer, sehingga customer merasa layanan yang digunakan lebih lengkap dan worth it dibanding kompetitor.


### 3. Perkuat loyalty dan referral program

Fitur **numberofreferrals** menunjukkan bahwa customer dengan jumlah referral rendah cenderung lebih berisiko churn, sedangkan customer yang memiliki referral lebih tinggi cenderung menunjukkan loyalitas yang lebih kuat.

**Recommendation:**
Perusahaan dapat memperkuat program referral atau loyalty reward, misalnya memberikan benefit kepada customer yang berhasil mengajak pengguna baru. Program ini tidak hanya membantu akuisisi customer baru, tetapi juga meningkatkan engagement dan loyalitas customer lama.


### 4. Fokus pada customer yang terdeteksi churn oleh model

Model yang telah dibuat dapat digunakan untuk membantu tim retention menentukan customer mana yang perlu diprioritaskan. Dengan threshold hasil optimization sebesar **0.30302092**, model menjadi lebih sensitif dalam mendeteksi customer yang berpotensi churn.

**Recommendation:**
Customer dengan probabilitas churn di atas threshold tersebut dapat dimasukkan ke dalam daftar prioritas retention campaign. Tim bisnis dapat membuat treatment bertingkat, misalnya:

1. Risiko tinggi: personal call, promo khusus, atau service recovery.
2. Risiko sedang: email campaign, loyalty offer, atau bundling service.
3. Risiko rendah: monitoring rutin tanpa intervensi besar.


## Model Limitation & Recommendation

Meskipun Best Model menunjukkan performa yang baik dengan F2-score test sebesar **93.56%** dan berhasil menurunkan False Negative dari **21 menjadi 8**, model ini masih memiliki beberapa batasan. Model masih menghasilkan False Positive yang lebih tinggi, yaitu meningkat dari **62 menjadi 94**, sehingga terdapat customer yang sebenarnya tidak churn tetapi tetap masuk ke target retention. Selain itu, threshold **0.30302092** bersifat business-driven dan perlu dievaluasi ulang jika strategi atau budget retention berubah.

Model juga masih bergantung pada data historis, sehingga performanya dapat menurun jika terjadi perubahan perilaku customer, strategi kompetitor, atau kondisi pasar. Selain itu, model belum mempertimbangkan faktor eksternal seperti promosi kompetitor, kualitas jaringan real-time, dan riwayat komplain terbaru.

Oleh karena itu, model direkomendasikan untuk digunakan sebagai **decision support tool** bagi tim retention, bukan sebagai satu-satunya dasar keputusan. Oleh karena itu, model perlu dimonitor secara berkala, threshold perlu disesuaikan dengan kebutuhan bisnis, dan data tambahan seperti customer complaint, campaign response, dan service quality dapat ditambahkan untuk meningkatkan akurasi dan relevansi model.
